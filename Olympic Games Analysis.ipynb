```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
#load the data sets:
df1=pd.read_csv('C:/Users/SOLOMON/Downloads/archive (6)/athlete_events.csv')
df2=pd.read_csv('C:/Users/SOLOMON/Downloads/archive (6)/noc_regions.csv')
```

```python
#merge the two dataframes:
df3=df1.merge(df2,how='left',on='NOC')
df3.head()
```

```python
#check number off rows and columns.
df3.shape
```

```python
#change the name of the colmuns to make them consistent.
df3.rename(columns={'region':'Region','notes':'Notes'})
```

```python
#top ten countries:
df3.groupby('Team').size().sort_values(ascending=False).head(10).plot(kind='bar',color=sns.color_palette('viridis'))
```

```python
#a hsitogram to show Age distribution of the partcipants:
plt.figure(figsize=(10,4))
plt.title('AGE DISTRIBUTION OF THE PARTCIPANTS')
plt.xlabel('Age')
plt.ylabel('NUMBER OF PARTCIPANTS')
plt.hist(df3.Age,bins=np.arange(15,55,2),color='orange',edgecolor='white');
```

```python
#games played in the summmer:
df3[df3.Season=='Winter'].Sport.unique()
```

```python
#total number of each gender:
GENDER=df3['Sex'].value_counts()
GENDER
```

```python
#pie chart representation of each Gender:gender disribution:
plt.figure(figsize=(10,6))

plt.pie(GENDER,labels=GENDER.index,autopct='%1.1f%%',startangle=100,shadow=True)
plt.show()
```

```python
#total numbers of medals:
medals=df3['Medal'].value_counts()
medals
```

```python
#piechart for the above:
plt.figure(figsize=(10,5))
#plt.title('MEDALS REPRESENTATION',fontsize=20)
plt.pie(medals,labels=medals.index,autopct='%1.1f%%',shadow=True);
```
